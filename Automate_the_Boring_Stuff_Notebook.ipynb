### Keyword Arguments:

- By default the print function has a new line /n character after it.
- the end= is a keyword argument
- if you dont want the next print statemnt to be on the new line you can specify the end= keyword argument e.g:
print("Hello", end='')

- By default the print statement prints a space between each argument that has been passed. eg. print("cat", "mouse", "dog") will print cat mouse dog. Each character is separated by a space character automatically.
- But I can change that too by using sep= keyword argument:
print("cat", "mouse", "dog", sep="ABC")
catABCmouseABCdog

#### Say for example I have a list and I want to print out all the elements in the list along with their index values. I can use range Function along with len function :

#### Python Multiple Assignment Trick:

```python
# say you have a list cat:

cat = ["fat", "orange", "loud"]

# now say you need to create a new variable to assign each of the attributes you would normally do this:

size = cat[0]
color = cat[1]
Type = cat[2]

print(size)
print(color)
print(Type)



# But you can also do the same with mutlple Assignments:

cat = ["fat", "orange", "loud"]

size1, color1, Type1 = cat

print(size1, color1, Type1)


size2, color2, Type2 = "skinny","black","quiet"

print(cat)
print(size2, color2,Type2)
```

5. **Sort Method:**

```python
spam = [15, 7.14, 0, -3.14 , 3.14, 2, 20]
print(spam)

spam.sort()
print(spam)

spam.sort(reverse=True)
print(spam)
```
