### List Refrencing

If you are referencing a list to another list and
If you change one element of one of the list, the same change will be updated in the other list as well





### Variables dont contain list, They only contain the refrences to the list

```python
spam = [0,1,2,3,4,5]
print(spam)

cheese = spam
print(cheese)

cheese[0] = "Hello"

print(cheese)
print(spam)
```

### Immutable values (like strings) dont have this problem because they cannot be modified "in place" theyncan only be replaced by new values


Now lets say for some reason in your previous code you dont want the cheede and spam to be referenced but act as individual lists such that changing the variables on one list should not affect the other, you can use the deepcopy() function as shown below:

```python
import copy

spam = [0,1,2,3,4,5]

cheese = copy.deepcopy(spam)

cheese[0] = "Hello"

print(spam)
print(cheese)
```

now you see that spam and cheese are two individual lists, seperate lists and not refrences to each other

### Lesson 17: Dictionary Data Type

Dictionary are data type with key, value pair

- Items in a Dictionay are unordered
- Dictionaries are **MUTABLE**

You can access keys, values and items in a dictionary like this:

```python
mycat = {'size':'fat','color':'gray','disposition ':'loud'}

print(mycat['size'])


print(mycat.keys())
print(mycat.values())
print(mycat.items())
```

**For loops with dictionaries**:

```python
mycat = {'size':'fat','color':'gray','disposition ':'loud'}

for k in mycat.keys():
    print(k)
    
print('-----------------------')
    
for v in mycat.values():
    print(v)

print('-----------------------')

for k,v in mycat.items():
    print(k,v)
```

setdefault() methods checks if a key exists in a dictionary and if not, it can also add that value to the dictionary

```python
shinobi = {'name':'naruto','jutsu':'shadow-clone',"clan":'Uzumaki'}

print(shinobi)

shinobi.setdefault('Tailed beast','nine tailed fox')

print(shinobi)
```

If say for example I want to change the 'Tailed beast' key to 'Shukaku' it wont let me do that because the key 'Tailed beast' is already defined.

### Write a program to count all the letters in a string

```python
import pprint

text = 'My name is Chinmay Rozekar. One day I will work at Google.'
count = {}


for each_character in text.lower():
    count.setdefault(each_character,0)
    count[each_character] +=1
    
some_text = pprint.pformat(count)
print(some_text)
    
```
